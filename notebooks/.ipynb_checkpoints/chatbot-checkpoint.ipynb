# Financial Chatbot

This chatbot answers predefined financial questions using company financial data.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/financial_dataset.csv")

# Preview data
df.head()
```

```python
def simple_chatbot(user_query):
    user_query = user_query.lower()

    # Microsoft Revenue
    if "revenue" in user_query and "microsoft" in user_query:
        value = df[df["Company"] == "Microsoft"]["Revenue (USD Billions)"].values[-1]
        return f"Microsoft's revenue is {value} billion USD."

    # Microsoft Liabilities
    elif "liabilities" in user_query and "microsoft" in user_query:
        value = df[df["Company"] == "Microsoft"]["Total Liabilities (USD Billions)"].values[-1]
        return f"Microsoft's total liabilities are {value} billion USD."

    # Apple Net Income Change
    elif "net income" in user_query and "apple" in user_query:
        apple_data = df[df["Company"] == "Apple"]
        change = apple_data["Net Income (USD Billions)"].values[-1] - apple_data["Net Income (USD Billions)"].values[-2]
        return f"Apple's net income changed by {change} billion USD."

    # Highest Revenue Company
    elif "highest revenue" in user_query:
        top = df.sort_values("Revenue (USD Billions)", ascending=False).iloc[0]
        return f"{top['Company']} has the highest revenue of {top['Revenue (USD Billions)']} billion USD."

    # Tesla Revenue
    elif "revenue" in user_query and "tesla" in user_query:
        value = df[df["Company"] == "Tesla"]["Revenue (USD Billions)"].values[-1]
        return f"Tesla's revenue is {value} billion USD."

    else:
        return "Sorry, I can only answer predefined financial questions."
```

```python
simple_chatbot("What is the revenue of Microsoft")
```

```python
simple_chatbot("What are Microsoft's liabilities")
```

```python
simple_chatbot("What is the revenue of Mirosoft")
```

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot ended.")
        break

    print("Bot:", simple_chatbot(user_input))

```python
## Chatbot Documentation

### Overview
This project is a rule-based financial chatbot that answers predefined financial queries using company financial data.

### Supported Queries
- What is Microsoft’s revenue?
- What are Microsoft’s liabilities?
- What is Apple’s net income change?
- Which company has the highest revenue?
- What is Tesla’s revenue?

### How it Works
The chatbot uses if-else logic to match user queries and retrieve relevant data from the dataset.

### Limitations
- Only supports predefined queries
- Cannot understand incorrect spelling
- No real-time data
```
